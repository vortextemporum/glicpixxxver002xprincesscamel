```python
from PIL import Image
from image4layer import Image4Layer
import os
import random

glicfolder = "/path/to/glicpixfolder/"
camelfolder = "/path/to/camelfolder/"

glicimages =  os.listdir(glicfolder)

camelimages = os.listdir(camelfolder)


a = Image.open(f'{glicfolder}{random.sample(glicimages, 1)[0]}')
b = Image.open(f'{camelfolder}{random.sample(camelimages, 1)[0]}')

a = a.resize((1024,1024),Image.NEAREST)
b = b.resize((1024,1024))

coin = random.randint(0,1)
print(coin)

if coin == 0:
    out = Image4Layer.exclusion(a, b)
if coin == 1:
    out = Image4Layer.exclusion(b,a)
out
```

```python
for i in range(250):
    a = Image.open(f'{glicfolder}{random.sample(glicimages, 1)[0]}')
    b = Image.open(f'{camelfolder}{random.sample(camelimages, 1)[0]}')

    a = a.resize((1024,1024),Image.NEAREST)
    b = b.resize((1024,1024))
    out1 = Image4Layer.exclusion(a, b)
    out1.save(f'./ab/{str(i).zfill(3)}.png')
    out2 = Image4Layer.exclusion(b, a)
    out2.save(f'./ba/{str(i).zfill(3)}.png')    
```
